```python
##maximun calculation
def maxf(a,b,f,n):
    x=np.linspace(0,1,n);rs=np.array([np.complex(np.cos(z*2*np.pi),np.sin(z*2*np.pi)) for z in x])
    m=np.abs(f(a,b,rs)).max()
    return m
##engine calculation
def engine(a,b,n):
    l=len(a)
    e=[np.abs(np.array([b[k]*(a[k].conjugate()**i) for k in range(l)]).sum())**2 for i in range(n)]
    return np.array(e).sum()
```

```python
def f1(a,b,z):
    return b/(1-a.conjugate()*z)
def f2(a,b,z):
    return b[0]/(1-a[0].conjugate()*z)+b[1]/(1-a[1].conjugate()*z)
def f3(a,b,z):
    return b[0]/(1-a[0].conjugate()*z)+b[1]/(1-a[1].conjugate()*z)+b[2]/(1-a[2].conjugate()*z)
```

```python
def disc_to_square(z):
    ang=np.angle(z)
    if (ang>=np.pi/4 and ang<3*np.pi/4) or (ang>=-3*np.pi/4 and ang<-np.pi/4):
        w=z/np.abs(np.sin(ang))
    else:
        w=z/np.abs(np.cos(ang))
    return w
    
def square_to_disc(w):
    ang=np.angle(w)
    if (ang>=np.pi/4 and ang<3*np.pi/4) or (ang>=-3*np.pi/4 and ang<-np.pi/4):
        z=w*np.abs(np.sin(ang))
    else:
        z=w*np.abs(np.cos(ang))
    return z
```

```python
## Lipschtizs constant in Granule
def Ggranule(r,a,b,delta):
    ep1=r[0]+np.array([(np.complex(1,1))*delta,(np.complex(1,-1))*delta,(np.complex(-1,1))*delta,-(np.complex(1,1))*delta])
    rmax1=np.array([np.abs(np.real(ep1)).max(),np.abs(np.imag(ep1)).max()]).max()
    g11=np.abs(np.array([(a[0].conjugate()-square_to_disc(z).conjugate())/(1-a[0].conjugate()*square_to_disc(z)) for z in ep1])).max()
    g12=np.abs(np.array([(a[1].conjugate()-square_to_disc(z).conjugate())/(1-a[1].conjugate()*square_to_disc(z)) for z in ep1])).max()
    #print(g11,g12)
    ep2=r[1]+np.array([(np.complex(1,1))*delta,(np.complex(1,-1))*delta,(np.complex(-1,1))*delta,-(np.complex(1,1))*delta])
    rmax2=np.array([np.abs(np.real(ep2)).max(),np.abs(np.imag(ep2)).max()]).max()
    g21=np.abs(np.array([(a[0].conjugate()-square_to_disc(z).conjugate())/(1-a[0].conjugate()*square_to_disc(z)) for z in ep2])).max()
    g22=np.abs(np.array([(a[1].conjugate()-square_to_disc(z).conjugate())/(1-a[1].conjugate()*square_to_disc(z)) for z in ep2])).max()
    #print(g21,g22)
    G1=(b[0]*g21/(1-rmax1*np.abs(a[0]))+b[1]*g22/(1-rmax1*np.abs(a[1])))*(b[0]*g21*g11/(1-rmax1*np.abs(a[0]))+b[1]*g22*g12/(1-rmax1*np.abs(a[1])))
    G2=(b[0]*g21/(1-rmax2*np.abs(a[0]))+b[1]*g22/(1-rmax2*np.abs(a[1])))*(b[0]*g21*g11/(1-rmax2*np.abs(a[0]))+b[1]*g22*g12/(1-rmax2*np.abs(a[1])))
    #print(G1,G2)
    return ((G1**2+G2**2)**0.5)
```

```python
def f32(a,b,z):
    l=len(a)
    F=[b[k]*(a[k].conjugate()-z[0].conjugate())/((1-a[k].conjugate()*z[1])*(1-a[k].conjugate()*z[0])) for k in range(l)]
    return np.array(F).sum()
def f33(a,b,z):
    l=len(a)
    F=[b[k]*(a[k].conjugate()-z[0].conjugate())*(a[k].conjugate()-z[1].conjugate())
       /((1-a[k].conjugate()*z[2])*(1-a[k].conjugate()*z[1])*(1-a[k].conjugate()*z[0])) for k in range(l)]
    return np.array(F).sum()
def e3(a,b,z):
    #print(z)
    E1=(1-np.abs(z[0])**2)*(np.abs(f3(a,b,z[0]))**2)
    E2=(1-np.abs(z[1])**2)*(np.abs(f32(a,b,z))**2)
    E3=(1-np.abs(z[2])**2)*(np.abs(f33(a,b,z))**2)
    return E1+E2+E3
```
